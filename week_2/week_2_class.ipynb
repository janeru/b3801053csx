## 利用BeautifulSoup爬取10頁的Ptt KoreaDrama版 

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

#### 需要注意的地方:
1.先找出一頁，而每個r-rent裡面又包含，推文、日期、作者、標題、標題連結，所以在分別處理，而在爬取超連結可能會因為文章有被刪除而沒有連結可用，所以用個try...except來處理

2.之後再把他們都放到list裡面，變成一項一項的

3.之後再處理多頁的部分

```python
url = 'https://www.ptt.cc/bbs/KoreaDrama/index.html'

info_dict=[]
infopush = []
infodate = []
infoauthor = []
infotitlehref = []
infotitle = []
#抓10頁
for round in range(10):
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
    articles = soup.find_all('div', 'r-ent')
    for article in articles:
        #推文
        push = article.select('.nrec')[0].text
        #日期
        date =  article.select('.date')[0].text
        #作者
        author = article.select('.author')[0].text
        #標題連結
        try:
            titlehref = article.select('.title')[0].find('a').get('href')
        except:
            titlehref = "null"
            next
        #標題
        title = article.select('.title')[0].text.strip()
        
        infopush.append(push)
        infodate.append(date)
        infoauthor.append(author)
        infotitlehref.append('https://www.ptt.cc'+titlehref)
        infotitle.append(title)
    #拿上一頁的連結
    paging = soup.select('div.btn-group-paging a')
    #print(paging)
    #做成下一頁的網址連結
    next_url = 'https://www.ptt.cc'+ paging[1]['href']
    url = next_url
```

#### 把大家再放入一個list

```python
info_dict = {'push' : infopush,
             'date' : infodate,
            'author': infoauthor,
            'href':infotitlehref,
            'title':infotitle}
```

轉成dataframe

```python
newform_df = pd.DataFrame(info_dict)
```

```python
newform_df.to_csv('ptt_koreadrama_info.csv', encoding='utf-8', index=None)
```
