## 3.資料分組(group_by)+繪圖

## 變項：購買商品數量隨月份分佈 的情形

```python
data1 = df.groupby(by='new_DateId')['Quantity'].sum()
```

```python
%matplotlib inline
b = data1.plot()
```

## 變項：買就送贈品 的隨月份的分佈情況

```python
data = df.groupby(by='new_DateId')['IsGift'].sum()
%matplotlib inline
a = data.plot()
```

## 小結：發現在11月份的時候，贈品的贈額達到最高峰，而購買商品的數量也最多，故推測贈品會刺激人們消費的數量

## 看PayProfileTypeDef付款方式的分佈

```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.countplot(x = "PayProfileTypeDef", data=df)
```

## 發現SevenEleven是消費者比較喜歡的貨到付款的方式

## 付款方式與單品價格的關係

```python
sns.factorplot(x = "PayProfileTypeDef", y = "UnitPrice", data = df, ci = None)
```

## 發現單品價格越高，賣家比較頃向希望消費者以ATM先轉帳後再出貨
