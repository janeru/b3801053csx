## 3.資料分組(group_by)+繪圖+進一步了解資料

### *變項：購買商品數量隨日期分佈 的情形

```python
data1 = df.groupby(by='new_DateId')['Quantity'].sum()
```

```python
%matplotlib inline
b = data1.plot()
```

### *變項：買就送贈品 的隨日期的分佈情況

```python
data = df.groupby(by='new_DateId')['IsGift'].sum()
%matplotlib inline
a = data.plot()
```

### 結果：發現在11月份的時候，贈品的贈額達到最高峰，而購買商品的數量也最多，應該是因為11月份有雙11購物節

### *變項：付款方式與單品價格的關係

```python
import seaborn as sns
sns.factorplot(x = "PayProfileTypeDef", y = "UnitPrice", data = df, ci = None)
```

### 結果：發現單品價格越高，賣家比較頃向希望消費者以ATM先轉帳後再出貨，這樣也比較可以減少出貨買家沒有拿貨的情況發生

### *看變項 PayProfileTypeDef付款方式的分佈(分成前後半年看)

```python
#選取2017年前半年的資料
d = df[df.new_DateId.between('2017-01-01','2017-06-30')]
#選取2017年後半年的資料
d1 = df[df.new_DateId.between('2017-07-01','2017-12-31')]
```

```python
%matplotlib inline
import seaborn as sns
sns.countplot(x = "PayProfileTypeDef", data=d)
```

```python
sns.countplot(x = "PayProfileTypeDef", data=d1)
```

### 結果：發現不論是前半年或後半年，SevenEleven是消費者比較常選擇的貨到付款的方式

### *TrackSourceTypeDef流量來源的變項討論

```python
dd = d[['TrackSourceTypeDef','new_DateId']]
pd.value_counts(dd.TrackSourceTypeDef)
```

```python
dd1 = d1[['TrackSourceTypeDef','new_DateId']]
pd.value_counts(dd1.TrackSourceTypeDef)
```

### 結果：發現不論是前半年或後半年，網路的流量都是以網站取勝，但後半年相較於前半年，AndroidApp的流量相較iosAPP的流量增加

### 看2017年的流量來源在各月份的分佈情況，因為覺得官網或是手機app的設計改變有可能會影響消費者喜好使用的情形

```python
df11 = df[df.new_DateId.between('2017-01-01','2017-12-31')]
```

```python
#把月份分離出來
#method1:df11['Month'] = df11['new_DateId'].apply(lambda x: x.strftime('%m')) #月份
#method2:
from datetime import datetime
#dt = datetime.strptime(a , "%Y-%m-%d').strftime('%m')
df11['month'] = df11['new_DateId'].dt.strftime('%m')
df11.tail(2)
```

```python
TST = df11.groupby('TrackSourceTypeDef').size()
```

```python
(df11.groupby(['month', 'TrackSourceTypeDef']).size().unstack().plot.bar())
```

### 結果：發現1,2月份其實Android app的與ios差不多，但3月以後明顯的都是ios比較多
