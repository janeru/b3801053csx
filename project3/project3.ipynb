## Step1:清資料、填補缺失值

先將年份有空值的整筆資料清除掉

```python
df = df[df['Year'].notnull()]
```

將有空的值，填補上0

```python
df.fillna(0, inplace=True)
```

```python
df.head()
```

## Step2:相關係數

想要看各個變數間每年份的變化-->會用差距去看相關係數

```python
df_diff = df.shift(-1)-df.shift(0) 
```

```python
df_diff.head()
```

去掉年份，再做相關係數

```python
df_diff=df_diff.drop('Year',axis=1)
```

```python
df_diff.corr()
```

變項重新命名

```python
df.rename(columns={'Prof.': 'Professor', 'AssProf.': 'AssProfessor','AsstProf.':'AsstProfessor','Prof.Outside':'ProfessorOutside','AssProf.Outside':'AssProfessorOutside','AsstProf.Outside':'AsstProfessorOutside'}, inplace=True)
```

```python
df.head()
```

## Step4:利用linear regression來對變項進行分析

由前面看出正教授與合聘講師的相關係數<0


```python
from statsmodels.formula.api import ols
model = ols("Professor ~Lecture", df).fit()
```

```python
model.summary()
```

正教授與合聘講師的關係成反比，但由p-value < 0.05故拒絕虛無假說，表示正教授的數量增加，但是合聘講師的數目並未減少

再看看正教授數量與講師在外面授課的情形，前面得到相關係數<0

```python
model = ols("Professor ~LectureOutside", df).fit()
model.summary()
```

```python
藉由linear regression看出，p-value<0.05，故拒絕虛無假說，因此無法得證，正教授數量增加，講師在外面授課的數量會減少
```

最後再藉由模型看正教授數量與各個變項之間的關係

```python
from statsmodels.formula.api import ols
model = ols('Professor ~Year+AssProfessor+AsstProfessor+Lecture+ProfessorOutside+AssProfessorOutside+AsstProfessorOutside+LectureOutside', df
    ).fit()
```

```python
model.summary()
```

由模型與各變項看起，看出教授數量與年分、外聘教授的相關係數>0外，與其他的變項相關係數都 < 0，但是與p-value的關係，在年份、助理教授、助理教授的授課數都是 < 0.05，因此拒絕虛無假說，並無法說聘正教授的數量會隨年份增加、或正教授在外授課的數量增加、或副教授在外授課的數量增加
